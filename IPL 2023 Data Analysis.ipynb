# Exploring the Impact of Winning a Toss based on IPL Matches

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
init_df=pd.read_csv('ipl-dataset/ipl_2023.csv')
init_df.head()
```

## Toss Decisions based on Stadiums

### Ahmedabad

```python
ahm1=len(init_df[(init_df["toss_decision"]=='field') & (init_df["location"]=="Ahmedabad")])
ahm2=len(init_df[(init_df["toss_decision"]=='bat') & (init_df["location"]=="Ahmedabad")])
labels = ['Field First', 'Bat First']
sizes = [ahm1, ahm2]
colors = ['lightblue', 'lightgreen']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Ahmedabad')
plt.axis('equal')
plt.show()
```
